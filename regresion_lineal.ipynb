```python
#Importar las librerias necesarias al entorno de trabajo

import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Cargar la data 

datos = pd.read_csv ("C:/Users/Jorge/Documents/datasets/regresion_lineal.csv", sep=",")
```

```python
# Explorar los datos cargados

datos.columns
datos.info
datos.describe
```

```python
# Visualizar los datos contenidos en el dataset

datos
```

```python
# Generar la gr√°fica de dispersi√≥n para los datos originales

datos.plot.scatter (x="metro", y="precio", color="lightseagreen")
plt.ylabel("Precio")
plt.xlabel("Metro")
plt.grid(axis = 'both', color = 'lightgrey', linestyle = 'dashed')
plt.show()
```

```python
# Creador de modelos de regresi√≥n lineal

regresion = linear_model.LinearRegression()
```

```python
# Agregar los datos dentro de un vector 

metros = datos["metro"].values.reshape((-1,1))
```

```python
# Crear el modelo de regresi√≥n lineal

modelo = regresion.fit(metros, datos["precio"])
```

```python
# Imprimir el punto de intersecci√≥n y la pendiente de la recta

print ("Intersecci√≥n(b)", modelo.intercept_)
print ("Pendiente (m)", modelo.coef_)
```

```python
# Definir entradas para el modelo de predicci√≥n e imprimir las predicciones

entradas = [[4],[5],[8],[10],[12],[15],[16],[20],[25]]
predicciones = modelo.predict(entradas)
print (predicciones)
```

```python
# Imprimir la grafica con el modelo de regresi√≥n lineal datos originales vs predicciones

datos.plot.scatter(x="metro", y="precio", label="Datos base", color ='lightcoral')
plt.scatter(entradas, predicciones, color="firebrick")
plt.plot (entradas, predicciones, color ="grey", label="Linea de regresi√≥n")
plt.xlabel ('metros')
plt.ylabel ('precio')
plt.legend()
plt.grid(axis = 'both', color = 'lightgrey', linestyle = 'dashed')
plt.show()
```
